# Data Visualization Notebook

This notebook focuses on data visualization and analysis related to credit scores.

## Import Libraries

The following libraries are essential for data manipulation and visualization.

```python
# Import necessary libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

## Data Loading and Preprocessing

Load the dataset and perform initial data cleaning.

```python
# Load the dataset and perform initial data cleaning.
df = pd.read_csv('credit_score.csv')
print(df.shape)
print(df.info())
string_columns = df.select_dtypes(include=['object']).columns
print(len(string_columns))
df = df.drop(['Name','Customer_ID', 'ID',"SSN"], axis=1)
```

## Data Summary and Visualization

Generate summary statistics and create various visualizations to understand the dataset.

```python
# Generate summary statistics and create various visualizations to understand the dataset.
df_first = df.iloc[:,0:12]
df_second = df.iloc[:,12:24]

summary_stats = df_first.describe().drop('count')
# Generate summary statistics of the data

summary_stats_second = df_second.describe().drop('count')
# Generate summary statistics of the data

# Create a point plot
plt.figure(figsize=(28, 8))
sns.heatmap(data=summary_stats, annot=True, fmt=".2f", cmap="viridis")
# Create a heatmap to visualize summary statistics
plt.title('Summary Statistics Heatmap')
plt.xticks()
plt.show()

plt.figure(figsize=(28, 8))
sns.heatmap(data=summary_stats_second, annot=True, fmt=".2f", cmap="viridis")
# Create a heatmap to visualize summary statistics
plt.title('Summary Statistics Heatmap')
plt.xticks()
plt.show()
```

```python
value_counts = df['Type_of_Loan'].value_counts()
# Count the occurrences of unique values in a column

# Remove rows with "No Data"
top_n = 10
other_count = value_counts[top_n:].sum()
top_value_counts = value_counts[:top_n]
top_value_counts['Other'] = other_count
# Aggregate less frequent values into 'Other'

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_value_counts, labels=top_value_counts.index, autopct='%1.1f%%', startangle=140)
# Create a pie chart to visualize the distribution of top loan types
plt.title('Distribution of Top Unique Loan Types')
plt.axis('equal')
plt.show()
```

```python
value_counts_df = value_counts.reset_index()
value_counts_df.columns = ['Type_of_Loan', 'Count']
# Rename columns for clarity

# Create a treemap using plotly
fig = px.treemap(value_counts_df, path=['Type_of_Loan'], values='Count', title='Treemap of Loan Types')
# Create an interactive treemap to visualize loan type distribution
fig.show()
```

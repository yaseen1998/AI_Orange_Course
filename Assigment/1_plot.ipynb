```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

```python
df = pd.read_csv('../csv/credit_score.csv')
df = df.drop(['Name','Customer_ID', 'ID',"SSN"], axis=1)
```

```python
df_first = df.iloc[:,0:12]
df_second = df.iloc[:,12:24]

summary_stats = df_first.describe().drop('count')
summary_stats_second = df_second.describe().drop('count')

# Create a point plot
plt.figure(figsize=(28, 8))
sns.heatmap(data=summary_stats, annot=True, fmt=".2f", cmap="viridis")
plt.title('Summary Statistics Heatmap')
plt.xticks()
plt.show()

plt.figure(figsize=(28, 8))
sns.heatmap(data=summary_stats_second, annot=True, fmt=".2f", cmap="viridis")
plt.title('Summary Statistics Heatmap')
plt.xticks()
plt.show()
```

```python
value_counts = df['Type_of_Loan'].value_counts()

# remove all row No Data  
top_n = 10
other_count = value_counts[top_n:].sum()
top_value_counts = value_counts[:top_n]
top_value_counts['Other'] = other_count

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_value_counts, labels=top_value_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Top Unique Loan Types')

plt.axis('equal')
plt.show()
```

```python
value_counts = df['Type_of_Loan'].value_counts()
value_counts_df = value_counts.reset_index()
value_counts_df.columns = ['Type_of_Loan', 'Count']

# Create a treemap using plotly
fig = px.treemap(value_counts_df, 
                 path=['Type_of_Loan'], 
                 values='Count', 
                 title='Treemap of Loan Types')

# Show the interactive treemap
fig.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
def hitogram_string_column(df):
    fig,axes = plt.subplots(5,1,figsize=(30,40))
    columns = df.columns
    ax_row = 0
    for column in columns:
            sns.histplot(x=df[column],ax=axes[ax_row])
            axes[ax_row].set_title(column)
            ax_row += 1
            
    plt.show()
```

```python
df = pd.read_csv("../csv/credit_score.csv")
df = df.drop(['Name','Customer_ID', 'ID',"SSN"], axis=1)
x = df.loc[:, ['Occupation',"Credit_Mix",'Payment_of_Min_Amount','Payment_Behaviour','Credit_Score']]
print(x.head())
hitogram_string_column(x)
```

```python
value_counts = df['Type_of_Loan'].value_counts()

# remove all row No Data  
top_n = 10
other_count = value_counts[top_n:].sum()
top_value_counts = value_counts[:top_n]
top_value_counts['Other'] = other_count

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_value_counts, labels=top_value_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Top Unique Loan Types')

plt.axis('equal')
plt.show()
```

```python
value_counts = df['Type_of_Loan'].value_counts()
value_counts_df = value_counts.reset_index()
value_counts_df.columns = ['Type_of_Loan', 'Count']

# Create a treemap using plotly
fig = px.treemap(value_counts_df, 
                 path=['Type_of_Loan'], 
                 values='Count', 
                 title='Treemap of Loan Types')

# Show the interactive treemap
fig.show()
```

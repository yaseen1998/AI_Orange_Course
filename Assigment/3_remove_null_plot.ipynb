```python
import plotly.express as px
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df2 = pd.read_csv("remove_null.csv")
```

```python

x = df2['Payment_of_Min_Amount']
# Create a histogram of the data column and change the number of bins
plt.hist(x, bins=3, color='orange')

# Add labels and title
plt.xlabel('Payment of Min Amount')
plt.ylabel('Count')
plt.title('Histogram of Payment of Min Amount')

# Show the plot
plt.show()
```

```python
value_counts = df2['Type_of_Loan'].value_counts()

value_counts_df = value_counts.reset_index()
value_counts_df.columns = ['Type_of_Loan', 'Count']

# Create a treemap using plotly
fig = px.treemap(value_counts_df, 
                 path=['Type_of_Loan'], 
                 values='Count', 
                 title='Treemap of Loan Types')

# Show the interactive treemap
fig.show()
```

```python
value_counts = df2['Type_of_Loan'].value_counts()

# remove all row No Data  
top_n = 10
other_count = value_counts[top_n:].sum()
top_value_counts = value_counts[:top_n]
top_value_counts['Other'] = other_count

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_value_counts, labels=top_value_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Top Unique Loan Types')

plt.axis('equal')
plt.show()
```

```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import numpy as np
```

```python
df2 = pd.read_csv('../csv/assigment/credit_score_new.csv')
value_counts = df2['Type_of_Loan'].value_counts()

value_counts_df = value_counts.reset_index()
value_counts_df.columns = ['Type_of_Loan', 'Count']

# Create a treemap using plotly
fig = px.treemap(value_counts_df, 
                 path=['Type_of_Loan'], 
                 values='Count', 
                 title='Treemap of Loan Types')

# Show the interactive treemap
fig.show()
```

```python
x = df2['Payment_of_Min_Amount']
# Create a histogram of the data column and change the number of bins
plt.hist(x, bins=3)

# Add labels and title
plt.xlabel('Payment of Min Amount')
plt.ylabel('Count')
plt.title('Histogram of Payment of Min Amount')

# Show the plot
plt.show()
```

```python
loan = ['Personal Loan', 'Home Equity Loan', 'Auto Loan', 'Debt Consolidation Loan', 'Not Specified', 'Payday Loan', 'Student Loan', 'Mortgage Loan', 'Credit-Builder Loan']
df = pd.read_csv('../csv/assigment/credit_score_clean.csv')
x = df.loc[:, loan]



boolean_counts = x.apply(pd.Series.value_counts)

# Plot the counts using a bar chart
ax = boolean_counts.T.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Counts of True/False in Loan Types')
plt.xlabel('Loan Types')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Values', loc='upper right', labels=['False', 'True'])
plt.show()
```

```python
boolean_counts = x.apply(pd.Series.value_counts)

# Create a clustered bar chart
fig, ax = plt.subplots(figsize=(10, 6))
x = np.arange(len(loan))
width = 0.4

ax.bar(x - width/2, boolean_counts.loc[False], width, label='False')
ax.bar(x + width/2, boolean_counts.loc[True], width, label='True')

ax.set_title('Counts of True/False in Loan Types')
ax.set_xlabel('Loan Types')
ax.set_ylabel('Count')
ax.set_xticks(x)
ax.set_xticklabels(loan, rotation=45, ha='right')
ax.legend(title='Values', loc='upper right')

plt.tight_layout()
plt.show()
```

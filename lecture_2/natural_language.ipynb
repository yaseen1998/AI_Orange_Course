```python
# Import spacy 
import spacy
```

```python
# Use (spacy.load()) to load the english pipeline
nlp = spacy.load(name='en_core_web_sm')
```

```python
# Read the wiki_us.txt file using with open()
with open('../csv/wiki_us.txt') as f:
    doc = nlp(f.read())
    
```

```python
# Use (nlp()) to create an nlp object from the text 
nlp = nlp(doc.text)
print(nlp)
```

```python
# Loop through first 10 tokens in doc
for token in doc[:10]:
    # Print the token and its part-of-speech tag
    print(token, token.pos_)
```

```python
# Print the third sentence in the document
third_sentence = list(doc.sents)[2]
print(third_sentence)
```

```python
# Print ORG entites
orgs = [ent.text for ent in doc.ents if ent.label_ == 'ORG']
print(orgs)
```
